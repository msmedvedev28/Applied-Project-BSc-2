#### Using the functions

```python
import time
import datetime
from config import PRODUCT_URL, PRODUCT_NAME, CHECK_INTERVAL_SECONDS
from database import init_database, save_price_to_db
from parser import get_current_price
from inflation import print_inflation_report

# Initialization
init_database()
print(f"Price monitoring: {PRODUCT_NAME}")
print(f"🔗 {PRODUCT_URL}")
print(f"⏱️ Interval: {CHECK_INTERVAL_SECONDS} сек\n")

# Infinite Loop
iteration = 0
try:
    while True:
        iteration += 1
        print(f"\n{'='*50}")
        print(f"🔁 Iteration #{iteration} - {datetime.datetime.now()}")
        
        price = get_current_price(PRODUCT_URL)
        if price:
            save_price_to_db(PRODUCT_NAME, price)
            print_inflation_report()
        else:
            print("⚠️ Price not found")
            
        print(f"💤 Next check after {CHECK_INTERVAL_SECONDS} сек")
        time.sleep(CHECK_INTERVAL_SECONDS)
        
except KeyboardInterrupt:
    print("\n🛑 Interrupted by the user")
```
